```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
calories = np.array([2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000])
exercise = np.array([60, 50, 30, 90, 45, 20, 75, 95, 15, 10])
bmi = np.array([25, 26, 28, 23, 27, 30, 24, 22, 31, 33])
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
# Daily Calorie Intake vs. BMI
axes[0].scatter(calories, bmi, color='b')
axes[0].set_xlabel("Daily Calorie Intake")
axes[0].set_ylabel("BMI")
axes[0].set_title("Daily Calorie Intake vs. BMI")
axes[0].grid()
# Exercise Duration vs. BMI
axes[1].scatter(exercise, bmi, color='g')
axes[1].set_xlabel("Exercise Duration (mins)")
axes[1].set_ylabel("BMI")
axes[1].set_title("Exercise Duration vs. BMI")
axes[1].grid()
# Daily Calorie Intake vs. Exercise Duration
axes[2].scatter(calories, exercise, color='r')
axes[2].set_xlabel("Daily Calorie Intake")
axes[2].set_ylabel("Exercise Duration (mins)")
axes[2].set_title("Daily Calorie Intake vs. Exercise Duration")
axes[2].grid()
plt.tight_layout()
plt.show()
```

```python
# Pearson Correlation Coefficients
corr_calories_bmi, _ = pearsonr(calories, bmi)
corr_exercise_bmi, _ = pearsonr(exercise, bmi)
corr_calories_exercise, _ = pearsonr(calories, exercise)
print(f"Correlation between Daily Calorie Intake and BMI: {corr_calories_bmi:.2f}")
print(f"Correlation between Exercise Duration and BMI: {corr_exercise_bmi:.2f}")
print(f"Correlation between Daily Calorie Intake and Exercise Duration: {corr_calories_exercise:.2f}")
```

```python
if corr_calories_bmi > 0.8:
    print("Higher calorie intake is strongly associated with a higher BMI.")
if corr_exercise_bmi < -0.8:
    print("More exercise duration is strongly linked to a lower BMI.")
if corr_calories_exercise < -0.8:
    print("Higher calorie intake tends to be associated with less exercise.")
```
