Generate a normal distribution of data points with a mean of 60 and a standard
deviation of 12. Plot a histogram for the generated data. Analyze the symmetry and
bell-shaped curve of a normal distribution and visualize its Probability Density
Function (PDF).

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
data = np.random.normal(loc=60, scale=12, size=1000)
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')
x = np.linspace(min(data), max(data), 1000)
pdf = stats.norm.pdf(x, loc=60, scale=12)
plt.plot(x, pdf, 'r', label='PDF')
plt.title('Normal Distribution (Mean=60, SD=12)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

The normalized distribution for a person’s daily calorie intake has a mean of 30
calories and a standard deviation of 5 calories. Write a python code that Calculates and
visualizes the following probabilities for a randomly selected person having a daily
calorie intake:
I. Greater than or equal to 19 calories.
II. Less than or equal to 19 calories.
III. Between 19 and 35 calories.


```python
mean_calories = 30
std_calories = 5
p_ge_19 = 1 - stats.norm.cdf(19, loc=mean_calories, scale=std_calories)
p_le_19 = stats.norm.cdf(19, loc=mean_calories, scale=std_calories)
p_between_19_35 = stats.norm.cdf(35, loc=mean_calories, scale=std_calories) - stats.norm.cdf(19, loc=mean_calories, scale=std_calories)
print(f"P(X >= 19): {p_ge_19:.4f}")
print(f"P(X <= 19): {p_le_19:.4f}")
print(f"P(19 <= X <= 35): {p_between_19_35:.4f}")
x = np.linspace(mean_calories - 4 * std_calories, mean_calories + 4 * std_calories, 1000)
pdf = stats.norm.pdf(x, mean_calories, std_calories)
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, 'r', label='PDF')
plt.fill_between(x, pdf, where=(x >= 19), color='g', alpha=0.3, label='P(X >= 19)')
plt.fill_between(x, pdf, where=(x <= 19), color='b', alpha=0.3, label='P(X <= 19)')
plt.fill_between(x, pdf, where=((x >= 19) & (x <= 35)), color='yellow', alpha=0.5, label='P(19 <= X <= 35)')
plt.title('Probability Distribution of Daily Calorie Intake')
plt.xlabel('Calorie Intake')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```
